```python
from tensorflow.keras.models import load_model
import pandas as pd
```

# 1. Load & process data

```python
df = pd.read_csv('Dataset/kddcup.testdata.unlabeled', header=None)
df.head()
```

```python
df.info()
```

```python
# Kiểm tra duplicated row & drop 
print('No of duplicated row: ', df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print('New df.shape:',df.shape)
```

```python
# cột 0,4,5 vì các giá trị unique
df = df.drop([0,4,5], axis=1)
df.head()
```

```python
# Load pipeline
import pickle
with open('pipeline.pkl','rb') as file:
  pipeline = pickle.load(file)
```

```python
df=df.values
df = pipeline.transform(df.tolist())
```

# 2. Load model & Prediction

```python
# Load model
model = load_model('Q2_ANN_model.h5')
```

```python
# Predictions
df_pred = model.predict(df)
```

```python
# Làm tròn về giá trị 0 và 1
df_pred[:10].round(0)
```

```python
# Đưa về giá trị label ('normal.': 0,'neptune.': 1,'other.': 2)
df_pred.argmax(axis=1)[:10]
```
